```python
data = pd.read_fwf('Moon Distance df output.txt')
data.reset_index(drop=True).style
#print(data)
```

# Get targets observed >= 20 

```python
target_list = data['OBJECT'].value_counts()  # get counts of each target

#find which objects have min 20 observations
good_targets = target_list[(target_list >= 20)] 
#print(good_targets)

good_obs = np.zeros(len(data), dtype=bool) # creating empty bool array of zeros of same length as data frame

for obj in good_targets.keys():      # loop through every object name of good targets 
    good_obs = good_obs | (data['OBJECT']==obj)    # or matches, 1's for 

    
new_df = data[good_obs]

#print(new_df)
#print(np.sum(good_obs))
#new_df.reset_index(drop=True).style
```
